```python
import cv2
import numpy as np
from keras.models import model_from_json

# Cargar la arquitectura del modelo desde el archivo JSON
ruta_archivo_json = "modelo_mobilenetv2.json"
with open(ruta_archivo_json, "r") as archivo_json:
    modelo_cargado_json = archivo_json.read()

# Crear un nuevo modelo a partir de la arquitectura JSON cargada
modelo_cargado = model_from_json(modelo_cargado_json)

# Cargar los pesos en el modelo
modelo_cargado.load_weights("model_mobilenetv2.h5")
```

```python
IMG_SIZE = (224, 224)
CLASSES = ['papaya','orange','banana', 'aloevera']
```

```python

# Inicializar la cámara (0 indica la cámara predeterminada, puede ajustar según sea necesario)
cap = cv2.VideoCapture(0)

while True:
    # Capturar cada fotograma
    ret, frame = cap.read()

    # Redimensionar el fotograma al tamaño del modelo (IMG_SIZE)
    frame = cv2.resize(frame, IMG_SIZE)

    # Realizar la predicción en el fotograma
    input_array = np.expand_dims(frame, axis=0)  # Ajusta según las necesidades de tu modelo
    predictions = modelo_cargado.predict(input_array)

    # Obtener la clase predicha
    predicted_class = CLASSES[np.argmax(predictions)]

    # Mostrar la clase predicha en la ventana de la cámara
    cv2.putText(frame, predicted_class, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.imshow('Clasificación en Tiempo Real', frame)

    # Romper el bucle si se presiona la tecla 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Liberar la cámara y cerrar la ventana
cap.release()
cv2.destroyAllWindows()
```
